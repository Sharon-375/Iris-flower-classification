```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import geopandas as gpd

# Create your own custom dataset using a DataFrame
data = {
    'sepal length (cm)': [5.1, 4.9, 6.2, 5.8, 7.1, 6.4, 5.0, 6.3, 4.7, 5.5],
    'sepal width (cm)': [3.5, 3.0, 3.4, 2.7, 3.0, 2.8, 3.6, 3.3, 3.2, 2.5],
    'petal length (cm)': [1.4, 1.4, 5.4, 4.1, 5.9, 4.5, 1.5, 5.6, 1.3, 4.0],
    'petal width (cm)': [0.2, 0.2, 2.3, 1.0, 2.1, 1.5, 0.2, 2.4, 0.2, 1.3],
    'species': ['setosa', 'setosa', 'virginica', 'versicolor', 'virginica',
                'versicolor', 'setosa', 'virginica', 'setosa', 'versicolor']
}

# Create the DataFrame
iris = pd.DataFrame(data)
iris['species'] = pd.Categorical(iris['species'])

# Set seaborn style
sns.set(style="whitegrid")

# 1. Histogram: Sepal Length
plt.figure(figsize=(6, 4))
sns.histplot(data=iris, x='sepal length (cm)', hue='species', kde=True)
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()

# 2. Bar Plot: Species count
plt.figure(figsize=(6, 4))
sns.countplot(data=iris, x='species', palette='Set2')
plt.title("Bar Plot of Species Count")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()

# 3. Box Plot: Petal Width by Species
plt.figure(figsize=(6, 4))
sns.boxplot(data=iris, x='species', y='petal width (cm)', palette='Pastel1')
plt.title("Box Plot of Petal Width")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.show()

# 4. Scatter Plot: Sepal vs Petal Length
plt.figure(figsize=(6, 4))
sns.scatterplot(data=iris, x='sepal length (cm)', y='petal length (cm)', hue='species', s=70)
plt.title("Scatter Plot: Sepal vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()

# 5. Heat Map: Correlation Matrix
plt.figure(figsize=(6, 5))
corr = iris.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heat Map of Feature Correlations")
plt.show()

# 6. Map Visualization (Dummy World Map using GeoPandas)
#world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
#world.plot(color='lightblue', edgecolor='black')
#plt.title("World Map Visualization (Example)")
#plt.show()

```
